```python
import speech_recognition as sr
from bnunicodenormalizer import Normalizer
import pandas as pd
from tqdm import tqdm
from glob import glob
import os
import librosa
tqdm.pandas()
norm=Normalizer()
```

```python
def infer(AUDIO_FILE):
    # use the audio file as the audio source
    r = sr.Recognizer()
    with sr.AudioFile(AUDIO_FILE) as source:
        audio = r.record(source)  # read the entire audio file
    text = r.recognize_google(audio, language ='bn-BD')
    text = str(text)
    return text
#     words=text.split()
#     text=[norm(w)["normalized"] for w in words]
#     return " ".join(text)
```

```python
files = os.listdir(data_dir)
print(files)
print()
print(len(files))
```

```python
info = []

for i in files:
    aud =  f"{data_dir}{i}"
    audDuo = []
    audDuo.append(i)
    try:
        pred=infer(aud)
        audDuo.append(pred)
    except Exception as e:
        audDuo.append("<>")
        #print(e)

    info.append(audDuo)
```
