# Bangla Wav2Vec2-large

```python
import requests

API_URL = "https://api-inference.huggingface.co/models/auditi41/wav2vec2-large-xlsr-53-Bangla"
headers = {"Authorization": "Bearer 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"}

def query(filename):
    with open(filename, "rb") as f:
        data = f.read()
    response = requests.post(API_URL, headers=headers, data=data)
    return response.json()
```

```python
external_id = []
contents = []
errors = []
c = 0
#c&i = []

def once(output,i):

    external_id.append(i)

    if len(output['text']) == 0:
        contents.append("<>")
    else:
        contents.append(output['text'])


def again(audio,i):

    global c

    try:
        output = query(audio)

        if 'error' in output.keys():
            if c != 5:
                c += 1
                again(audio)
            else:
                c = 0
                external_id.append(i)
                contents.append("<>")
                pass
        else:
            once(output,i)


    except Exception as e:
        again(audio,i)


for i in x:

    audio = f"{data_dir}{i}"


    try:
        output = query(audio,i)
        once(output)

    except:
        again(audio,i)
```
